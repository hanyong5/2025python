- genres 에서 [] 데이터는 제외

    movies['genres'].fillna('[]') \
- 문자열을 리스트데이터로 변환

    .apply(literal_eval) \
- 장르 데이터를 추출하여 리스트로 변경

    .apply(lambda x: sorted(i['name'] for i in x) if isinstance(x, list) else [] ) \
- 리스트데이터를 하나의 문자열로 변환

    .apply(lambda x:" ".join(x) if len(x)>0 else None )

```python
# movies['genres'].eq('[]').value_counts()

# movies['genres'].fillna('[]').apply(literal_eval).apply(lambda x: 참 if 조건 else 거짓 )
# movies['genres'].fillna('[]').apply(literal_eval).apply(lambda x: sorted(표현식 for i in x) if isinstance(x, list) else [] )
# test_list = movies['genres'].fillna('[]') \
#     .apply(literal_eval) \
#     .apply(lambda x: sorted(i['name'] for i in x) if isinstance(x, list) else [] )

movies['str_genres'] = movies['genres'] \
    .apply(literal_eval) \
    .apply(lambda x: sorted(i['name'] for i in x) if isinstance(x, list) else [] ) \
    .apply(lambda x:" ".join(x) if len(x)>0 else None )

```

### BoW기반 Contents Based Filtering



```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
bow_vector = CountVectorizer()
```

```python
genre_mat = bow_vector.fit_transform(movies['str_genres'])
```

```python
dict(sorted(bow_vector.vocabulary_.items()))
```

```python
type(genre_mat)
genre_mat.toarray()[0]
```

### 코사인 유사도 측정

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_of_genre = cosine_similarity(genre_mat,genre_mat) # data:genre_mat
```
