```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fruits= np.load('data/fruits_300.npy')
```

```python
fruits.shape
```

```python
fruits[0,0,:]
```

```python
plt.imshow(fruits[0],cmap='gray')
plt.show()
```

```python
fig,axs= plt.subplots(2,3)
axs[0,0].imshow(fruits[0],cmap='gray')
axs[0,1].imshow(fruits[100],cmap='gray_r')
axs[0,2].imshow(fruits[200],cmap='gray_r')
axs[1,0].imshow(fruits[50],cmap='gray_r')
axs[1,1].imshow(fruits[150],cmap='gray_r')
axs[1,2].imshow(fruits[250],cmap='gray_r')
plt.show()
```

```python
apple = fruits[0:100].reshape(100,100*100)
pineapple = fruits[100:200].reshape(-1,100*100) # 100,100,100
banana = fruits[200:300].reshape(-1,100*100)
```

```python
apple.shape
```

```python
apple.mean(axis=1)
```

```python
plt.hist(np.mean(apple,axis=1),alpha=0.5)
plt.hist(np.mean(pineapple,axis=1),alpha=0.5)
plt.hist(np.mean(banana,axis=1),alpha=0.5)
plt.legend(['apple','pineapple','banana'])
plt.show()
```

```python
fruits1 = np.array([
    [[1,2],
     [3,4]], #image 0
    [[2,2],
     [2,2]], #image 1
    [[4,4],
     [4,4]] #image 2
])


print(fruits1[0].mean(axis=1))
print(fruits1[1].mean(axis=1))
print(fruits1[2].mean(axis=1))
```

```python
fig,axs = plt.subplots(1,3,figsize=(20,5))
axs[0].bar(range(10000),np.mean(apple,axis=0))
axs[1].bar(range(10000),np.mean(pineapple,axis=0))
axs[2].bar(range(10000),np.mean(banana,axis=0))
plt.show()
```

```python
np.mean(apple, axis=0).shape
```

```python
apple_mean = np.mean(apple, axis=0).reshape(100,100)
pineapple_mean = np.mean(pineapple, axis=0).reshape(100,100)
banana_mean = np.mean(banana, axis=0).reshape(100,100)

fig,axs= plt.subplots(1,3)
axs[0].imshow(apple_mean,cmap='gray')
axs[1].imshow(pineapple_mean,cmap='gray_r')
axs[2].imshow(banana_mean,cmap='gray_r')
plt.show()
```
